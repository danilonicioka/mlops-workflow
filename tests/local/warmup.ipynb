# Average

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def extract_and_plot_average(directory):
    """
    Extracts the 'Average' column from all CSV files in the given directory
    and plots the values in a graph with differentiation for each file.
    
    Args:
        directory (str): Path to the directory containing CSV files.
    """
    plt.figure(figsize=(12, 8))  # Create a larger figure for better visualization

    # Iterate through all files in the directory
    for file in os.listdir(directory):
        if file.endswith(".csv"):
            file_path = os.path.join(directory, file)
            try:
                # Read the CSV file
                df = pd.read_csv(file_path)

                # Check if 'Average' column exists
                if 'Average' in df.columns:
                    # Plot the data with a label for the file name
                    plt.plot(
                        df['Average'].tolist(),
                        marker='o',  # Marker for points
                        linestyle='-',  # Line style
                        label=file  # Legend label for the file
                    )
                else:
                    print(f"'Average' column not found in {file}")
            except Exception as e:
                print(f"Error reading {file}: {e}")

    # Add graph details
    plt.title("Average Values from Multiple CSV Files", fontsize=16)
    plt.xlabel("Index", fontsize=14)
    plt.ylabel("Average Value", fontsize=14)
    plt.legend(title="File Source", loc="best", fontsize=12)  # Add legend
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
directory_path = "results/csv"
extract_and_plot_average(directory_path)
```

# Requests and Replicas

```python
# Load the data into a pandas DataFrame
file_path = 'test1_P2_I1.csv'
df = pd.read_csv(file_path)

# Extract the columns representing response times (all columns except the first and last)
response_columns = df.columns[1:-1]  # Exclude the first (replica) and last (average) columns
response_values = df[response_columns]

# Plot each request's response time (excluding the replica and average columns)
plt.figure(figsize=(10, 6))
for replica in range(30):
    plt.plot(range(1, 1 + len(response_values.iloc[replica])), response_values.iloc[replica], marker='o', label=f'Replica {replica+1}')

# Add labels and title
plt.xlabel('Request')
plt.ylabel('Response Time')
plt.title('Response Time for Each Request and Replica')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
